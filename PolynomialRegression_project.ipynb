Generate dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('w2_position_salaries.csv')
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(df['Level'],df['Salary'])
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
x = df["Level"].values
y = df["Salary"].values

x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

**Train a Simple Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
lin_reg_1=LinearRegression()
lin_reg_1.fit(x,y)
print("Degree 1 - R2 score : ",lin_reg_1.score(x,y))
```

**Train Polynomial Regression Model**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regr=PolynomialFeatures(degree=2)
X_poly2=poly_regr.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly2,y)
```

```python
X_poly_val2=poly_regr.fit_transform(x)
print("Degree 2 - R2 score : ",lin_reg_2.score(X_poly2,y))
```

```python
poly_regr=PolynomialFeatures(degree=3)
X_poly3=poly_regr.fit_transform(x)
lin_reg_3=LinearRegression()
lin_reg_3.fit(X_poly3,y)
```

```python
X_poly_val3=poly_regr.fit_transform(x)
print("Degree 3 - R2 score : ",lin_reg_3.score(X_poly3,y))
```

```python
poly_regr=PolynomialFeatures(degree=4)
X_poly4=poly_regr.fit_transform(x)
lin_reg_4=LinearRegression()
lin_reg_4.fit(X_poly4,y)
```

```python
X_poly_val4=poly_regr.fit_transform(x)
print("Degree 4 - R2 score : ",lin_reg_4.score(X_poly4,y))
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_1.predict(x),color='blue')

x2,y2=zip(*sorted(zip(x,lin_reg_2.predict(X_poly_val2)),key=lambda x: x[0]))
plt.plot(x2,y2,color='green')

x3,y3=zip(*sorted(zip(x,lin_reg_3.predict(X_poly_val3)),key=lambda x: x[0]))
plt.plot(x3,y3,color='black')

x4,y4=zip(*sorted(zip(x,lin_reg_4.predict(X_poly_val4)),key=lambda x: x[0]))
plt.plot(x4,y4,color='brown')

plt.title('Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
